```python
import numpy as np
np.random.seed(1)

import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

弱相關

```python
#隨機產生兩組介於1000的50個整數
x = np.random.randint(0, 50 ,1000)
y = np.random.randint(0, 50 ,1000)

#呼叫numpy 裡的相關矩陣函數
np.corrcoef(x, y)
```

```python
#劃出分布圖
plt.scatter(x,y)
```

正相關

```python
#隨機產生兩組介於1000的50個整數
x = np.random.randint(0, 50 ,1000)

y = x + np.random.randint(0, 10 ,1000)

#呼叫numpy 裡的相關矩陣函數
np.corrcoef(x, y)
```

```python
#劃出分布圖
plt.scatter(x,y)
```

負相關

```python
#隨機產生兩組介於1000的50個整數
x = np.random.randint(0, 50 ,1000)

y = -x + np.random.randint(0, 10 ,1000)

#呼叫numpy 裡的相關矩陣函數
np.corrcoef(x, y)
```

```python
#劃出分布圖
plt.scatter(x,y)
```
